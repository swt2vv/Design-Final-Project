## Final Product: The Electric Vehicles Revolution

My final product is an infographic which showcases what I believe to be some of the most prevelant trends and statistics of the dataset EV Adoption US, sourced through Kaggle (as further described in the README.ipynb). I chose this topic because over the summer I did research with UVA's Parking and Transportation department on EV adoption, legistics in procurring infrastructure, and the environmental impact of EV adoption in order to determine whether UVA should invest their move and encourage UVA staff, faculty, and vistors to drive their EVs to Grounds through installing charging stations. Since that experience, I have continued to pursue my curiousity in this subject through independent study in the news and I thought that this project was a perfect opportunity for me to drive deeper into the national trends that are associated with EV adoption. <br>

Through my research, I have found general trends, both state and regional level, of the relationship of EV registrations and several other factors (e.g. price per kWh, income, government-issued incentives). Below is the infographic summarizing some of my findings. 

![Final Product](graphs/FinalProduct.png)


The information placed on the top right are three interesting statistics I found when first analyzing this dataset. The most noteable statistics for me is that the total growth found in Electric Vehicle (EV) registrations is a 1,169.3% increase from 2016 to 2023, starting at 280 thousand to over 3.5 million. This dramatic increase is very compelling as it shows that the EV market has transformed rapidly from a niche industry to booming market of high-end vehicles. I further explore this increase in the graphic, "Growth of EV Registrations Overtime", in the middle left of the infographic which breaks down by year the number of EV registrations in the US and shows the calculated percent increase to the right of it. As you can see, this boom has been both consistently positive and high every single year from 2016 to 2023, and noteably 2022 has had the highest percent increase at 67.9%, meaning over double the number of EV registrations from 2021 to 2022.<br> <br>

After seeign this breakdown, I then wanted to better understand where these EV registrations are most prevelant. In the graph in the middle right, "Average EV Percentage Across States", I showcase the distribution of the EVs in each state by averaging over the years their EV percentage using the "EV Share (%)". As shown, it is apparent that the EV percentage of the western most region of the US has on average more high than the rest of the country. Most noteably, California had the highest average EV percentage over the years, at 1.46%. From solely this information, I theorized some potential factors that may be contributing to this distict difference such as number of wealthy individuals, political ideology, and more funding towards EV infrastructure i.e. charging stations. <br> <br>

To test out my theories, I looked into the other columns of this data set which collected information on the support for reduced taxes related to environmental policies ("reducetax"), average income per person ("Per_Cap_Income"), presence of state-level EV incentives ("Incentives"), number of public EV charging stations ("Stations"), and average price of electricity per kilwayy-hour ("price_cents_per_kwh"). I first created several scatterplots with correlation slopes of the relationship between each of these factors with the increase in EV registrations. I did this while grouping the states into regions (e.g. West South Central) which were determined by Wikipedia's "List of regions of the United States" page. I then used that calculated correlation value of each factor in for their regions and placed it on a horizontal bar graph to compare the factors correlation to EV registrations, which is the graph on the bottom right named "Correlation of Factors and EV Registrations by Region". From this visualization, I found it most interesting that reduced tax support has had a negative correlation with EV registrations in several regions such as the West South Central, Mountain, and East South Central regions. I further understand possible reasons why this may be occuring, I would need to research more about the type of support and taxation incentived that each state in these regions are offering. Additionally, in the regions where it had a positive correlation, the correlation is consistantly the weakest of the other factors which begs the question on the effectiveness these policies and support for reduced taxes has on EV adoption. Is it because the reduced taxes are minimal, or are there other factors effecting this correlation. <br> <br>

Lastly, I explored one of the factors with the highest correlation value to EV registrations which was charging stations. There are three main charging stations for EVs: Level one, Level two, and Level three (or also known as DC-Fast charging). In this pie chart on the bottom left of the infographic, "Distribution of Charging Stations in 2023", I simply explored the distribution of these stations of the total 176,032 which are installed and in use as of 2023. I am not surprised to see that Level two charging stations are the most common charging stationed, followed by DC-Fast chargers, and then Level one. Although DC-Fast charging stations are the most efficient and quickest charging station, it is much more expensive and, at times, not necessary for where a corperation wants to install a public charging station. For example, at UVA, staff, faculty, and visitors are the people that would primarily use charging stations installed on Grounds. Level two chargers charger an EV from 20% to 80% in 4-8 hours, which is the perfect amount of time for a person to charge their car until either their lunch break or until the end of a shift. Also, visitor too may want to charge their car to full after travelling to UVA, and could leave their car for several hours while they visit Grounds. This applies to several other corporations for work, hotels, and other services. Therefore, Level two charging is the most cost effective choice for many seeking to install public chargers. <br> <br>

Overall, from this dataset, we can further understand the relationship between several factors with the increase in EV registrations. Although it is difficult to determine whether it is the factor or EV registrations that is effecting the other, it is still important and intriguing to find these correlations and use this information to make changes in policy and encourage public outreach to see is these correlations are proven to show casuality through several trials.
